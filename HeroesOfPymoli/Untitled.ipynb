```python
#IMPORT DEPENDENCIES
import pandas as pd
import numpy as np

#STORE FILEPATH IN VARIABLE
purchase_data = "Resources/purchase_data.csv"

#READ DATA FILE WITH PANDAS LIBRARY
purchase_data_df = pd.read_csv(purchase_data)

purchase_data_df.head()
```

```python
#----------PLAYER COUNT----------
#FIND TOTAL NUMBER OF NON REPEATING PLAYERS
all_players = len(purchase_data_df["SN"].unique())

#DISPLAY DATA IN NEW DATAFRAME
total_players_df = pd.DataFrame({"Total Players": [all_players]})

total_players_df
```

```python
#----------PURCHASING ANALYSIS (TOTAL)----------
#FIND NUMBER OF UNIQUE ITEMS
item_list = len(purchase_data_df["Item ID"].unique())

#AVERAGE PURCHASE PRICE
average_price = round(purchase_data_df["Price"].mean(),2)

#TOTAL NUMBER OF PURUCHASES
total_transactions = len(purchase_data_df["Purchase ID"].unique())

#TOTAL REVENUE
total_revenue = round(sum(purchase_data_df["Price"]),2)

total_revenue = '{:,.2f}'.format(total_revenue)

#PLACE ALL NEW VALUBLES IN A SUMMARY DATAFRAME
purchase_analysis_df = pd.DataFrame({"Number of Unique Items": item_list,
                                     "Average Price": "$" + str(average_price),
                                     "Number of Purchases": [total_transactions],
                                     "Total Revenue": "$" + str(total_revenue)})
purchase_analysis_df
```

```python
#----------GENDER DEMOGRAPHICS----------
#FIND GENDER COUNT AND GENDER COUNT PERCENTAGE FOR EACH NON-REPEATING SN
#group by gender per unique sn
gender_count = purchase_data_df.groupby("Gender").SN.nunique()
gender_counts_df = pd.DataFrame(purchase_data_df.groupby("Gender").SN.nunique())
gender_counts_df = gender_counts_df.rename(columns={"SN": "Total Count"})

gender_counts_df["Perentage of Players"] = (round(gender_counts_df["Total Count"] / sum(gender_counts_df["Total Count"])*100,2)).map("{:.2f}%".format)

gender_counts_df.sort_values(["Total Count"], ascending=False)
```

```python
#----------PURCHASING ANALYSIS (GENDER)----------
#GROUPBY GENDER
gendf = purchase_data_df.groupby(["Gender"])
#PURCHASE COUNT BY GENDER 
###purchaseCount = gendf["Purchase ID"].count()
#AVERAGE PURCHASE PRICE BY GENDER
###avgPrice = round(gendf["Price"].mean(),2)
#TOTAL PURCHASE VALUE BY GENDER
###totalValue = gendf["Price"].sum()
#TOTAL PURCHASE AVERAGE PER EACH PERSON OF EACH GENDER
##purchaseAvg = round((totalValue / gender_count),2) 

genderAnalysisdf = pd.DataFrame({
    "Purchase Count": gendf["Purchase ID"].count(),
    "Average Purchase": "$" + str(round(gendf["Price"].mean(),2)),
    "Total Purchase Value": "$" + str(gendf["Price"].sum()),
    "Avg Total Purchase Per Person": "$" + str(round((totalValue / gender_count),2))})
genderAnalysisdf
```

```python
#----------AGE DEMOGRAPHICS----------
# CREATE 4 YEAR RANGE FOR EACH AGE BIN AND THEN BIN NAMES
ageBins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 100]
ageGroups = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
#CATEGORIZE INTO BINS
purchase_data_df["Age Groups"] = pd.cut(purchase_data_df["Age"], ageBins, labels=ageGroups)

Age_Demographics_df = pd.DataFrame({
    "Total Count": purchase_data_df.groupby(["Age Groups"])["SN"].nunique(), 
    "Percentage of Players": round(ageCount / purchase_data_df["SN"].nunique() * 100,2).map("{:.2f}%".format)})
Age_Demographics_df
```

```python
#BIN INITIAL DF BY AGE
#CALCULATE PURCHASE COUNT
purchase_data_df.groupby(["Age Groups"])["Purchase ID"].count()
#AVERAGE PURCHASE PRICE
purchase_data_df.groupby(["Age Groups"])["Price"].mean()
#TOTAL PURCHASE PRICE
purchase_data_df.groupby(["Age Groups"])["Price"].sum()
#AVERAGE PURCHASE TOTAL PER PERSON
purchaseAvg = purchase_data_df.groupby(["Age Groups"])["Price"].sum() / purchase_data_df.groupby(["Age Groups"])["SN"].nunique()
#CREATE SUMMARY DF
age_anlysis_df = pd.DataFrame({"Purchase Count": purchase_data_df.groupby(["Age Groups"])["Purchase ID"].count(),
                   "Average Purchase Price": purchase_data_df.groupby(["Age Groups"])["Price"].mean(), 
                   "Total Purchase Value": purchase_data_df.groupby(["Age Groups"])["Price"].sum(),
                   "Avg Total Purchase per Person": purchaseAvg,})
age_anlysis_df

```

```python
#----------TOP SPENDERS----------
topFive = purchase_data_df.groupby(["SN"])
purchase_data_df.groupby(["SN"])["Purchase ID"].count()
purchase_data_df.groupby(["SN"])["Price"].mean()
purchase_data_df.groupby(["SN"])["Price"].sum()

#CREATE SUMMARY DF
top_spenders_df = pd.DataFrame({"Purchase Count": purchase_data_df.groupby(["SN"])["Purchase ID"].count(),  
                                "Average Purchase Price": purchase_data_df.groupby(["SN"])["Price"].mean(),
                                "Total Purchase Value": purchase_data_df.groupby(["SN"])["Price"].sum(),})
top_spenders_df.sort_values(["Total Purchase Value"], ascending=False)
```
